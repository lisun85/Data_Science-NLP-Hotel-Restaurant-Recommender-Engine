```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Creating tf-idf matrix
tfidf = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=5, max_df=0.9, stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_percent['reviews_list'].values.astype('U'))

tfidf_pd = pd.DataFrame(tfidf_matrix.toarray(), columns = tfidf.get_feature_names())
tfidf_pd
```

```python
from sklearn.decomposition import NMF

nmf = NMF(4, init = "nndsvda")
nmf.fit(tfidf_matrix)
nmf
```

```python
topic_term = nmf.components_.round(3)
topic_term.shape
```

```python
topic_term_df = pd.DataFrame(topic_term.round(3),
                index = ["component_1", "component_2", "component_3", "component_4"], #"component_5"], #"component_6", "component_7"],
                columns = tfidf.get_feature_names())
topic_term_df
```

```python
# Function to display the top n terms in each topic
def display_topics(model, feature_names, no_top_words, topic_names = None): 
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix + 1)
        else:
            print("\nTopic: ", topic_names[ix])
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
    print("\n")
    return model, feature_names, no_top_words
```

# Recommender System 2 - Using TF IDF & Cosine Similiarity Matrix

```python
indices = pd.Series(df_percent.index)
```

```python
indices
```

```python
# Creating tf-idf matrix
tfidf = TfidfVectorizer(analyzer='word', ngram_range=(1, 2), min_df=5, max_df=0.9, stop_words='english')
tfidf_matrix = tfidf.fit_transform(df_percent['reviews_list'].values.astype('U'))
```

```python
#create cosine_similiarities
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
print(cosine_similarities)
```

```python
#Testing different ways of defining index, and creating descending list of cosine simliarity vectors

idx = indices[indices == 'The Flying Squirrel'].index[0]
score_series = pd.Series(cosine_similarities[idx]).sort_values(ascending=False)
score_series

#idx = indices2['Punjabi Home Made Food']
#idx
```

```python
#checking score_series make sense
idx = indices[indices == 'The Flying Squirrel']
idx
```

```python
top = list(score_series.iloc[0:50].index)
top
```

```python
#Creating a list of recommended restuarants and appending the cosine similarity vectors to it
recommend_restaurant = []
for i in top:
    recommend_restaurant.append(list(df_percent.index)[i])
recommend_restaurant
```

```python
#Creating new data frame and populating it with recommended restaurants
df_new = pd.DataFrame(columns=['cuisines', "Mean Rating", 'cost'])

for i in recommend_restaurant:
        df_new = df_new.append(pd.DataFrame(df_percent[['cuisines','Mean Rating', 'cost']][df_percent.index == i].sample()))
        
df_new
```

```python
#Removing Dups and Sorting by Mean Rating
df_new = df_new.drop_duplicates(subset=['cuisines','Mean Rating', 'cost'], keep=False)
df_new = df_new.sort_values(by='Mean Rating', ascending=False)
df_new
```

```python
#Putting all the Functions together

def recommend(name, cosine_similarities = cosine_similarities):
    
    recommend_restaurant = []
    
    #Defining the index of the restaurant
    idx = indices[indices == name].index[0]
    
    #sorting Cosine Similiarities Matrix in Descending Order
    score_series = pd.Series(cosine_similarities[idx]).sort_values(ascending=False)
                             
    #Filtering out the top 45 restaurants                         
    top = list(score_series.iloc[0:50].index)
    
    
    #Create list of top 45 recommended restaurants
    for i in top:
        recommend_restaurant.append(list(df_percent.index)[i])
        
    #New Dat Frame
    df_new = pd.DataFrame(columns=['cuisines', "Mean Rating", 'cost'])
    
    #Append Resturant to new Data Frame
    for i in recommend_restaurant:
        df_new = df_new.append(pd.DataFrame(df_percent[['cuisines','Mean Rating', 'cost']][df_percent.index == i].sample()))
        
    ##Eliminate Dups and sort by Mean Rating to show only the top 10.     
    df_new = df_new.drop_duplicates(subset=['cuisines','Mean Rating', 'cost'], keep=False)
    df_new = df_new.sort_values(by='Mean Rating', ascending=False).head(10)

    print('Top 10 Restaurants with Similar Reviews:')
    
    return df_new
```

```python
#Bakery / Desert Recommendations
recommend("Cake Menu")
```

```python
#Cafe Recommendations
recommend("The Flying Squirrel")
```

```python
#Fast Food Recommendations
recommend("Sandwichwallas")
```

```python
#North Indian & Chinese Cuisine Recommendations
recommend("Spicy Paratha House")
```

```python
def recommend2(name, cosine_similarities, indices):
    
    recommend_restaurant = []
    
    #Get index of the restaurant that matches the input name
    idx = indices[indices == name].index[0]
    
    #Get the pairwise similarity scores
    sim_scores = list(enumerate(cosine_similarities[idx]))
    
    #Sort the restaurants based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse = True)
    
    #Get the scores for 10 most similiar restaurants
    sim_scores = sim_scores[1:11]
    
    #Get the resturant indices
    restaurant_indices = [i[0] for i in sim_scores]
    

    return df_percent2['name'].iloc[restaurant_indices]
```
